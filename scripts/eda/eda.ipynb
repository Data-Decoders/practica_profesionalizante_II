```python
import pandas as pd

def calculate_percentage(df, column_name):
    """
    Calcula y muestra los porcentajes de cada elemento en una columna específica.
    
    :param df: DataFrame con los datos.
    :param column_name: Nombre de la columna.
    """
    if column_name in df.columns:
        counts = df[column_name].value_counts()
        percentages = (counts / counts.sum()) * 100
        percentage_df = pd.DataFrame({
            'Count': counts,
            'Percentage': percentages
        })
        print(f"\nPorcentajes de los elementos en la columna '{column_name}':")
        print(percentage_df)
        return percentage_df
    else:
        print(f"La columna '{column_name}' no existe en el DataFrame.")
        return None

for column_name in  ["store", "manufacturer", "category", "createdAt", "gender"]:
    calculate_percentage(df, column_name)
```

```python
# ===========================
# Analizar una columna categórica
# ===========================
for column in ["store", "manufacturer", "category", "createdAt", "gender"]:
    categorical_column = column  # Reemplazar por una columna categórica del DataFrame

    if df is not None:
        analyze_categorical_data(df, categorical_column)
    else:
        print("El DataFrame no está disponible.")
```
